## Quiz: Chapter 3
1. **Question:**  WWhat is the difference between a spreadsheet and a database?
   **Answer:**  A spreadsheet and a database differ in the way they store data. Spreadsheets store data in rows and columns, making it easily accessible for individuals while a database organizes data into related tables that can be linked through relationships. This, in turn, allows for greater structure, scalability, multi-user access, and stronger data integrity overall.

2. **Question:** What are the advantages and disadvantages of using a spreadsheet for health data management?
   **Answer:** Advantages include low technical requirements, quick calculations, and simple visualization methods. Disadvantages, on the other hand, include higher risk of human error, weak data validation controls, and challenges managing large/multi-user databases.

3. **Question:**  What are the advantages and disadvantages of using a database for health data management?
   **Answer:** Advantages include data integrity, secure access control, and the ability for multiple user to collaborate. Disadvantages, on the other hand, include greater techincal complexity, setup requirements, and ongoing maintenance.

4. **Question:** What are some examples of tasks that can be performed using a spreadsheet?
   **Answer:** Calculations, sorting data, filtering data, creating charts, analyzing small datasets, tracking budgets...

5.  **Question:** What are some examples of tasks that can be performed using a database?
   **Answer:** Managing health records, storing patient information across multiple tables, running queries, tracking data...

6. **Question:** What are some examples of tasks that can be performed using both a spreadsheet and a database?
   **Answer:** Data entry, generating reports, sorting data, filtering data, exporting summarized results...

7. **Question:** What are some examples of tasks that can be performed using neither a spreadsheet nor a database?
   **Answer:** Tasks that require physical interaction like clinical judgement or giving a real-time diagnostic.

8. **Question:** What are some examples of tasks that can be performed using a spreadsheet but not a database?
   **Answer:** Spreadsheets are better for quick informal tasks such as rapidly formatting data for a presentation, creating simple charts or performing simple single-user calculations.

9. **Question:** What are some examples of tasks that can be performed using a database but not a spreadsheet?
   **Answer:** Databases can create relationships across linked tables, manage complex and large databases effectively or support multi-user access with role-based permissions.
