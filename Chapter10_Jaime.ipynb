## Python Packages

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

%config InlineBackend.figure_format = 'retina'
sns.set_style('whitegrid')
```

## Dataset

```python
df = pd.read_csv('heart_failure.csv')

df.head()
df.columns
```

## Task 1

```python
sns.scatterplot(
    data=df,
    x='platelets',
    y='serum_sodium',
    hue='diabetes'
).set(
    title='Platelets and Serum Sodium',
    xlabel='Platelets [platelets/mL]',
    ylabel='Serum sodium [mEq/L]'
);

plt.show()
```

## Task 2

```python
df['platelets'].min(), df['platelets'].max()
```

```python
sns.displot(
    df,
    y='platelets',
    bins=range(0, 850000, 50000)
).set(
    title='Platelet Distribution',
    ylabel='Platelets [platelets/mL]',
    xlabel='Frequency'
);

plt.show()
```

## Task 3

```python
sns.catplot(
    data=df,
    x='diabetes',
    hue='death',
    kind='count'
).set(
    title='Diabetes by Survival Group',
    xlabel='Diabetes',
    ylabel='Frequency'
);

plt.show()
```
