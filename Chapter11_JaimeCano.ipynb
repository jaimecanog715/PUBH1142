## Python Packages

```python
import numpy as np
from scipy import stats
```

## Task

```python
np.random.seed(28)

dBP = np.random.normal(85, 10, 100000)

dBP.mean(), dBP.std()
```

## Task

```python
np.random.seed(12)

sample = np.random.choice(dBP, 40)

sample
```

## Task

```python
sample_mean = sample.mean()
sample_std = sample.std(ddof=1)
standard_error = sample_std / np.sqrt(40)

sample_mean, sample_std, standard_error
```

## Task 

```python
c = stats.t.ppf(0.975, 39)

lower_bound = sample_mean - c * standard_error
upper_bound = sample_mean + c * standard_error

lower_bound, upper_bound
```

## Task 

```python
np.random.seed(12)

bootstrapped_means = np.array([
    np.random.choice(sample, 40, replace=True).mean()
    for _ in range(1000)
])

bootstrapped_means.sort()
```
