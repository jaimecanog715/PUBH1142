## Packages Used 

```python
import numpy as np
from pandas import DataFrame
from patsy import dmatrices
import statsmodels.api as sm
```

## Quiz

```python
# Seed
np.random.seed(12)

# Sample size
n = 100

# Create data
wcc = np.round(np.random.normal(15, 5, n), 0)
crp = (wcc * 2) + np.round(np.random.normal(0, 10, n), 0)
lipase = wcc + crp + np.round(np.random.normal(2, 10, n), 0)

# DataFrame
df = DataFrame({
    'WCC': wcc,
    'CRP': crp,
    'Lipase': lipase
})

# Regression setup
y, X = dmatrices('Lipase ~ WCC + CRP', data=df)

# Fit model
model = sm.OLS(y, X).fit()

# Show results
model.summary2()
```

**F-statistic:**
The F-statistic is big and its p-value is below 0.05, meaning that the  means  model is statistically significant. Thus, we reject the null hypothesis and conclude that at least one predictor (WCC or CRP) is associated with Lipase.

**Coefficients and p-values:**
The intercept is not statistically significant. This is because WCC is statistically significant at the 5% level, while CRP is highly statistically significant (p < 0.001). In this case, both variables contribute to predicting Lipase, with CRP having a stronger effect.

**R² value:**
The R² value is high (around 0.76), meaning that approximately 76% of the variation in Lipase is explained by WCC and CRP. This implies that the model fits the data well.
