```python
# 1. Import pandas
import pandas as pd

# 2. Read a CSV file
df = pd.read_csv('filename.csv')

# 3. Display the first 5 rows
df.head()

# 4. Mean of Age
df['Age'].mean()

# 5. Median of Salary
df['Salary'].median()

# 6. Standard deviation of Score
df['Score'].std()

# 7. Number of missing values in each column
df.isnull().sum()

# 8. Correlation between Age and Salary
df['Age'].corr(df['Salary'])

# 9. Subset where Age > 30
df[df['Age'] > 30]

# 10. Range of Score
df['Score'].max() - df['Score'].min()

# 11. Group by Department and calculate mean Salary
df.groupby('Department')['Salary'].mean()

# 12. Group by Department and Job Title, then count rows
df.groupby(['Department', 'Job Title']).size()

# 13. Maximum Age in each Department
df.groupby('Department')['Age'].max()

# 14. Cross-tabulation: frequency count of Department and Job Title
pd.crosstab(df['Department'], df['Job Title'])

# 15. Cross-tabulation: mean Salary for each Department and Job Title combination
pd.crosstab(df['Department'], df['Job Title'], values=df['Salary'], aggfunc='mean')
```
