### Task 3 — Variable naming rules (PEP 8)

Key PEP 8 naming conventions:
- Use **snake_case** for variable names (e.g., `student_age`, `total_votes`).
- Variable names should be **descriptive** and lowercase.
- Avoid single-letter names except for simple counters (like `i`, `j`) in short loops.
- Constants are typically written in **UPPER_CASE** (e.g., `MAX_SIZE = 100`).
- Don’t use Python reserved keywords (like `for`, `if`, `while`, `class`) as variable names.
- Avoid confusing names like `l`, `O`, `I` (they look like 1 and 0).
